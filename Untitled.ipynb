```python
def get_frequent_itemsets_fast(transactions, min_support=0.02, max_len=2, min_item_count=50):

    n_transactions = len(transactions)
    
    # Convert transactions to sets (for quick subset operations)
    transactions = [set(t) for t in transactions]
    
    #  1. Count single items 
    item_support = {}
    for basket in transactions:
        for item in basket:
            item_support[item] = item_support.get(item, 0) + 1

    # 2. Filter out rare single items 
    frequent_itemsets = {
        frozenset([item]): count / n_transactions
        for item, count in item_support.items()
        if count >= min_item_count and count / n_transactions >= min_support
    }

    all_frequent = dict(frequent_itemsets)
    current_freq = list(frequent_itemsets.keys())
    k = 2

    # 3. Generate larger itemsets up to max_len 
    while current_freq and k <= max_len:
        # Candidate generation with pruning
        candidates = set()
        for i in range(len(current_freq)):
            for j in range(i + 1, len(current_freq)):
                union = current_freq[i].union(current_freq[j])
                if len(union) == k:
                    # Pruning: all (k−1)-subsets must be frequent
                    subsets = combinations(union, k - 1)
                    if all(frozenset(s) in all_frequent for s in subsets):
                        candidates.add(frozenset(union))
        
        if not candidates:
            break

        #  4. Count candidate supports 
        candidate_counts = {c: 0 for c in candidates}
        for basket in transactions:
            for candidate in candidates:
                if candidate.issubset(basket):
                    candidate_counts[candidate] += 1

        #  5. Keep only frequent itemsets 
        Lk = {
            c: count / n_transactions
            for c, count in candidate_counts.items()
            if count / n_transactions >= min_support
        }

        if not Lk:
            break

        all_frequent.update(Lk)
        current_freq = list(Lk.keys())
        k += 1

    #  6. Return as DataFrame 
    df_freq = pd.DataFrame([
        {'itemset': list(k), 'support': v} for k, v in all_frequent.items()
    ])
    df_freq = df_freq.sort_values(by='support', ascending=False).reset_index(drop=True)
    return df_freq


# Run the faster version
frequent_itemsets = get_frequent_itemsets_fast(
    transactions,
    min_support=0.02,      # increase support = faster run
    max_len=2,             # only pairs (you can set 3 later if needed)
    min_item_count=30      # ignore items appearing <30 times
)

print("Frequent itemsets found:", len(frequent_itemsets))
frequent_itemsets.head(10)
```

```python
def generate_association_rules(frequent_itemsets, min_confidence=0.3, min_lift=1.0):
    rules = []
    itemset_dict = {frozenset(i): s for i, s in zip(frequent_itemsets['itemset'], frequent_itemsets['support'])}
    
    for itemset in itemset_dict.keys():
        if len(itemset) >= 2:  # Only consider sets with 2+ items
            for item in itemset:
                antecedent = frozenset([item])
                consequent = itemset - antecedent
                
                support = itemset_dict[itemset]
                support_antecedent = itemset_dict.get(antecedent, 0)
                support_consequent = itemset_dict.get(consequent, 0)
                
                if support_antecedent > 0:
                    confidence = support / support_antecedent
                    if support_consequent > 0:
                        lift = confidence / support_consequent
                        if confidence >= min_confidence and lift >= min_lift:
                            rules.append({
                                'antecedent': list(antecedent),
                                'consequent': list(consequent),
                                'support': round(support, 4),
                                'confidence': round(confidence, 4),
                                'lift': round(lift, 4)
                            })
    
    df_rules = pd.DataFrame(rules)
    df_rules = df_rules.sort_values(by='lift', ascending=False).reset_index(drop=True)
    return df_rules

# Generate rules
association_rules = generate_association_rules(frequent_itemsets, min_confidence=0.4, min_lift=1.2)

print("Total rules generated:", len(association_rules))
association_rules.head(10)
```

```python
# Top rules by confidence
top_confidence = association_rules.sort_values(by="confidence", ascending=False).head(10)
print("🔹 Top 10 rules by confidence:\n")
print(top_confidence[['antecedent', 'consequent', 'support', 'confidence', 'lift']])

# Top rules by lift
top_lift = association_rules.sort_values(by="lift", ascending=False).head(10)
print("\n🔹 Top 10 rules by lift:\n")
print(top_lift[['antecedent', 'consequent', 'support', 'confidence', 'lift']])
```
